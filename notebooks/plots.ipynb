```python
import matplotlib.pyplot as plt
import numpy as np

# MAPE values for each model (unscaled and scaled)
mape_unscaled = {
    "ARIMA": 0.86611,
    "LSTM": 19.93629,
    "GRU": 19.57284
}

mape_scaled = {
    "ARIMA": 1.15486,
    "LSTM": 0.11797,
    "GRU": 0.11169
}

# Improved colors for the plots
colors_unscaled = ['#FF9999', '#66B2FF', '#99FF99']  # Using distinct and vibrant colors
colors_scaled = ['#FFCC99', '#99CCFF', '#CCFF99']  # Light and distinct colors

# Create plots

# 1. MAPE of all models (Unscaled)
plt.figure(figsize=(7,5))
plt.bar(mape_unscaled.keys(), mape_unscaled.values(), color=colors_unscaled)
plt.title('MAPE Comparison of Models (Unscaled Data)')
plt.ylabel('MAPE (%)')
plt.xlabel('Models')
plt.show()

# 2. MAPE of all models (Scaled)
plt.figure(figsize=(7,5))
plt.bar(mape_scaled.keys(), mape_scaled.values(), color=colors_scaled)
plt.title('MAPE Comparison of Models (Scaled Data)')
plt.ylabel('MAPE (%)')
plt.xlabel('Models')
plt.show()

# 3. Comparison Plots of MAPE for ARIMA, LSTM, and GRU (Scaled vs Unscaled)
# ARIMA comparison (Scaled vs Unscaled)
plt.figure(figsize=(7,5))
arima_mape = [mape_unscaled["ARIMA"], mape_scaled["ARIMA"]]
plt.bar(["Unscaled", "Scaled"], arima_mape, color=['#ADD8E6', '#C0C0C0'])  # Swapped colors
plt.title('MAPE Comparison of ARIMA (Scaled vs Unscaled)')
plt.ylabel('MAPE (%)')
plt.show()

# LSTM comparison (Scaled vs Unscaled)
plt.figure(figsize=(7,5))
lstm_mape = [mape_unscaled["LSTM"], mape_scaled["LSTM"]]
plt.bar(["Unscaled", "Scaled"], lstm_mape, color=['#87CEEB', '#D3D3D3'])  # Swapped colors
plt.title('MAPE Comparison of LSTM (Scaled vs Unscaled)')
plt.ylabel('MAPE (%)')
plt.show()

# GRU comparison (Scaled vs Unscaled)
plt.figure(figsize=(7,5))
gru_mape = [mape_unscaled["GRU"], mape_scaled["GRU"]]
plt.bar(["Unscaled", "Scaled"], gru_mape, color=['#A4D3EE', '#BEBEBE'])  # Swapped colors
plt.title('MAPE Comparison of GRU (Scaled vs Unscaled)')
plt.ylabel('MAPE (%)')
plt.show()
```
